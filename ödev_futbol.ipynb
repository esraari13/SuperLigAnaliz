```python
#en ykek dzeltilmi puanla sampiyon olan takim

maxpuan = lig["puan_duzeltilmis"].max()
lig[lig["puan_duzeltilmis"]==maxpuan]
```

```python
#en dsk puanla sampiyon olan takim
minpuan = lig["puan_duzeltilmis"].min()
lig[lig["puan_duzeltilmis"]==minpuan]
```

```python
#En yukaek puan ortalamas ile sampiyon olan takim 
lig["ort_puan"]=lig["puan_duzeltilmis"]/lig["oyun"]
maxortpuan=lig["ort_puan"].max()
lig[lig["ort_puan"]==maxortpuan]
```

```python
#En az puan ortalamas ile sampiyon olan takm 

minortpuan=lig["ort_puan"].min()
lig[lig["ort_puan"]==minortpuan]
```

```python
#sampiyon olan takmlarn mac bas ortalama gol saylar
lig["ort_gol"]=lig["atilan_gol"]/lig["oyun"]

```
